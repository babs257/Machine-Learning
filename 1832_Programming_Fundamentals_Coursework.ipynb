```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
MyGraph = nx.Graph()
```

```python
#first Trainline - Central line
MyGraph.add_node('Holborn', npos=(10, 180), ccn='#483C32')
MyGraph.add_node('Chancery Lane', npos=(18, 180), ccn='#00FF00')
MyGraph.add_node('St. Pauls', npos=(36, 180), ccn='#00FF00')
MyGraph.add_node('Bank', npos=(54, 180), ccn='#00FF00')
MyGraph.add_node('Liverpool Street', npos=(72, 180), ccn='#00FF00')

#second line - Piccadilly line
MyGraph.add_node('Holborn', npos=(10, 145), ccn='#483C32')
MyGraph.add_node('Covent Garden', npos=(28, 145), ccn='#FFFF00')
MyGraph.add_node('Leicester Square', npos=(46, 145), ccn='#FFFF00')
MyGraph.add_node('Piccadilly Circus', npos=(64, 145), ccn='#FFFF00')
MyGraph.add_node('Green Park', npos=(72, 125), ccn='#483C32')


#third line -Jubilee line
MyGraph.add_node('St. Johns Wood', npos=(10, 108), ccn='#FF00FF')
MyGraph.add_node('Baker Street', npos=(28, 108), ccn='#FF00FF')
MyGraph.add_node('Bond Street', npos=(46, 108), ccn='#FF00FF')
MyGraph.add_node('Green Park', npos=(72, 108), ccn='#483C32')
MyGraph.add_node('Westminster', npos=(82, 98), ccn='#483C32')


#fourth line  - District line
MyGraph.add_node('Monument', npos=(10, 30), ccn='#FF0000')
MyGraph.add_node('Cannon Street', npos=(28, 30), ccn='#FF0000')
MyGraph.add_node('Mansion House', npos=(46, 30), ccn='#FF0000')
MyGraph.add_node('Blackfriars', npos=(64, 30), ccn='#FF0000')
MyGraph.add_node('Temple', npos=(46, 50), ccn='#483C32')

#fifth line - Circle line
MyGraph.add_node('Victoria', npos=(10, 72), ccn='#0000FF')
MyGraph.add_node('St. James Park', npos=(28, 72), ccn='#0000FF')
MyGraph.add_node('Westminster', npos=(46, 72), ccn='#483C32')
MyGraph.add_node('Embankment', npos=(64, 72), ccn='#0000FF')
MyGraph.add_node('Temple', npos=(82, 72), ccn='#483C32')
```

```python
#first egde
MyGraph.add_edge('Holborn', 'Chancery Lane', cce='#00FF00',Dist = '0.4Km')
MyGraph.add_edge('Chancery Lane', 'St. Pauls', cce='#00FF00',Dist = '1.03Km')
MyGraph.add_edge('St. Pauls', 'Bank', cce='#00FF00',Dist = '0.74Km')
MyGraph.add_edge('Bank', 'Liverpool Street', cce='#00FF00',Dist = '0.74Km')

#second edge
MyGraph.add_edge('Victoria', 'St. James Park', cce='#0000FF',Dist = '0.72Km')
MyGraph.add_edge('St. James Park', 'Westminster', cce='#0000FF',Dist = '0.76Km')
MyGraph.add_edge('Westminster', 'Embankment', cce='#0000FF',Dist = '0.69Km')
MyGraph.add_edge('Embankment', 'Temple', cce='#0000FF',Dist = '0.71Km')

#third edge
MyGraph.add_edge('Monument', 'Cannon Street', cce='#FF0000',Dist = '0.34Km')
MyGraph.add_edge('Cannon Street', 'Mansion House', cce='#FF0000',Dist = '0.31Km')
MyGraph.add_edge('Mansion House', 'Blackfriars', cce='#FF0000',Dist = '0.61Km')
MyGraph.add_edge('Blackfriars', 'Temple', cce='#FF0000',Dist = '0.76Km')

#FOURTH edge
MyGraph.add_edge('St. Johns Wood', 'Baker Street', cce='#FF00FF',Dist = '2.06Km')
MyGraph.add_edge('Baker Street', 'Bond Street', cce='#FF00FF',Dist = '1.66Km')
MyGraph.add_edge('Bond Street', 'Green Park', cce='#FF00FF',Dist = '1.37Km')
MyGraph.add_edge('Green Park', 'Westminster', cce='#FF00FF',Dist = '1.34Km')

#fifth edge
MyGraph.add_edge('Holborn', 'Covent Garden', cce='#FFFF00',Dist = '0.59Km')
MyGraph.add_edge('Covent Garden', 'Leicester Square', cce='#FFFF00',Dist = '0.25Km')
MyGraph.add_edge('Leicester Square', 'Piccadilly Circus', cce='#FFFF00',Dist = '0.49Km')
MyGraph.add_edge('Piccadilly Circus', 'Green Park', cce='#FFFF00',Dist = '0.56Km')
```

```python
# Now we extract attributes from the graph to dictionaries
pos = nx.get_node_attributes(MyGraph,'npos')
nodecolour = nx.get_node_attributes(MyGraph,'ccn')
edgecolour = nx.get_edge_attributes(MyGraph,'cce')
edge_labels = nx.get_edge_attributes(MyGraph,'Dist')

line_colors = {'Central Line': '#00FF00','Piccadilly Line': '#FFFF00', 'Jubilee Line': '#FF00FF','District Line': '#0000FF','Circle Line': '#FF0000'}
# Place the values from the dictionaries in lists
NodeList = list(nodecolour.values())
EdgeList= list(edgecolour.values())
plt.figure(figsize=(10,10))
plt.text(40, 170, s='Central Line', rotation=0)
plt.text(40, 20, s='Cirlcle Line', rotation=0)
plt.text(40, 65, s='District Line', rotation=0)
plt.text(40, 100, s='Jubilee Line', rotation=0)
plt.text(40, 140, s='Picadally Line', rotation=0)

# Draw the graph's nodes and edges

nx.draw_networkx(MyGraph, pos, node_color=NodeList, font_size = 8)
nx.draw_networkx_edges(MyGraph, pos, edge_color=EdgeList)
nx.draw_networkx_edge_labels(MyGraph, pos, edge_labels = edge_labels, font_size = 6)
#create legend
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', label=label, markerfacecolor=color, markersize=10) for label, color in line_colors.items()]
plt.legend(handles=legend_handles, loc='lower right')
# Visualise the graph
plt.show()
```
